# Gender Inequality Analysis

Gender inequality is the social phenomenon in which people are not treated equally on the basis of gender. The treatment may arise from distinctions regarding biology, psychology, or cultural norms prevalent in the society. Gender inequality remains a significant and persistent issue across various societies worldwide, affecting individuals in multiple aspects of their lives. From education and employment opportunities to economic empowerment and political representation, the disparities between genders continue to hinder social progress and equality. In order to address and combat these inequalities effectively, it is crucial to gain a comprehensive understanding of the extent and nature of gender disparities in different domains.

This data analysis project aims to explore and analyze gender inequality by examining various datasets and indicators related to education, employment, income, and representation. By delving into the available data, we seek to uncover patterns, trends, and key insights that can shed light on the existing gender disparities and their implications. The findings of this project will contribute to the growing body of knowledge on gender inequality, enabling policymakers, researchers, and advocates to make informed decisions and implement targeted interventions for promoting gender equality.

**Now let's look at the Gender Inequality Index**

**How is it defined?**

The Gender Inequality Index is a composite measure reflecting inequality in achievements between women and men in three dimensions: reproductive health, empowerment and the labour market.

> The health dimension is measured by the maternal mortality ratio and the adolescent fertility rate.
> The empowerment dimension is measured by the share of parliamentary seats held by each gender, and by secondary and higher education attainment levels.
> The labour dimension is measured by women's participation in the workforce.

The GII varies between 0 (when women and men fare equally) and 1 (when men or women fare poorly compared to the other in all dimensions). It is designed to reveal the extent to which national human development achievements are eroded by gender inequality, and to provide empirical foundations for policy analysis and advocacy efforts.

#### Q2: DOES HUMAN DEVELOPMENT HAVE A DIRECT RELATIONSHIP WITH GENDER INEQUALITY INDEX?
I intend to find out
> - If countries can attain gender equality with low human capital development.
> - Will I find a country with high human development but a low GII?

To achieve this I will categorize the GII column using the pandas cut function. 
This breaks numeric data into groups according to limits or boundries set called Bins. the data will be displayed in a new column which will be qualitative(categorical).
> - The limits I will be using are the quartile ranges of the GII column.

```python
# creating new column for the categorized data
inequality_df["Gender_inequality_Bins"] = pd.cut(
    x=inequality_df["Gender_Inequality_Index"],
    bins= [0.012000, 0.177500, 0.363000, 0.505750, 0.820000],
    labels =["Very Low", "Low","High","Very High"]
)
inequality_df.sample(10)
```

```python
# plot GII vs human development per country
sns.scatterplot(data = inequality_df, x ='Country', 
                y='Gender_Inequality_Index', 
                hue= inequality_df.Human_development, s=100).set(title='DISTRIBUTION OF GII PER COUNTRY');
```

From the chart, we can observe that there is a general trend of decreasing GII as human development index increases. we can see that each colour is clustered together indicating a direct relationship between the two variables. Countries with high human development have a relatively lower GII index and the lower the GII index the more gender equal a country will be. Therefore its safe to say that human development is a good indicator of gender equality. *The higher the human development the lower the gender inequality index. ceteris paribus*

Additionally, we can see that some countries with similar human development indices have significantly different levels of gender inequality, indicating that there are other factors at play.

Overall, this chart highlights the importance of considering multiple factors when analyzing gender inequality and its impact on human development.
